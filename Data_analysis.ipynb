# 実践的なデータ解析を行う
乳房の腫瘍を計測したデータにおいて、良性と悪性の場合に分け、それぞれの**radius_mean**に差があるのかを検定していきます。なお、有意水準は5％とします。

- radius_mean：中心から外周までの平均距離（半径）

>*帰無仮説（H0）：良性の腫瘍半径　=　悪性の腫瘍半径*

>*対立仮説（H1）：良性の腫瘍半径　<　悪性の腫瘍半径*

### 1. 必要なライブラリのインポート

```python
import numpy as np
import pandas as pd 
import scipy as sp
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import statsmodels.formula.api as smf
import statsmodels.api as sm

%matplotlib inline
```

### 2. データの読み込み

```python
# データを読み込んで変数 data に格納
data = pd.read_csv('breast_cancer_wisconsin_data.csv')
```

```python
# データの最初の5行だけ表示
data.head()
```

### 3. データの概要確認

```python
# 統計量の確認
data.describe()
```

```python
# 欠損値の確認
print(data.isnull().sum())
```

### 4. データの前処理

```python
# カテゴリ値の数値化
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])

# 数値化した状態を確認してみる
data.head()
```

```python
# 良性と悪性の場合の半径
radius1 = data[data['diagnosis'] == 0]['radius_mean']
radius2 = data[data['diagnosis'] == 1]['radius_mean']
```

```python
# それぞれの平均値を確認
print(radius1.mean())
print(radius2.mean())
```

```python
# radius1,radius2について Q-Qプロットを表示して正規性を確認　
fig = sm.qqplot(radius1, line = 's')
fig = sm.qqplot(radius2, line = 's')
```

### 5. 仮説検定

```python
# 対応のないｔ検定
stats.ttest_ind(radius1, radius2, equal_var=False)
```

```python
# 片側検定により
pvalue = 1.6844591259582747e-64
p = pvalue / 2
p
```

### 6. 検定結果

p値が0.05を下回っているので、**"有意差あり"**とみなすことができ、  
**「良性の腫瘍半径は悪性の腫瘍半径より小さい」**と判断できました。
